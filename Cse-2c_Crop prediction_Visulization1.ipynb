```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/home/user/Crop_Prediction.csv")
```

```python
df['temperature'].head(100).plot(kind='line')
plt.title("Temperature Trend")
plt.show()
```

```python
df[['temperature','humidity','rainfall']].head(100).plot(kind='line',figsize=(7,5),marker='h',title='Climate Factors Trend')
plt.grid(True)
plt.show()
```

```python
df['humidity'].head(1000).plot(kind='line',title="Humidity Trend",color='orange')
plt.show()
```

```python
df['crop'].value_counts().plot(kind='bar',color="green")
plt.title("Crop produced count")
plt.show()
```

```python
df['rainfall'].plot(kind='hist', bins=20,
    color='blue',
    edgecolor='violet',
    figsize=(6,4),
    xlabel="Rainfall",
    title="Rainfall Distribution")
```

```python
df['soil_moisture'].plot(kind='density',title='Soil Moisture Density',color="brown")
```

```python
df['ph'].head(1000).plot(kind='hist',title="Soil pH Distribution",color='purple')
plt.show()
```

```python
df[['nitrogen','phosphorus','potassium']].head(100).plot(kind='area',figsize=(8,5),title='NPK Nutrients')
plt.show()
```

```python
plt.bar(df['rainfall'],df['ph'],color='violet')
plt.xlabel("Crop Predicted")
plt.ylabel("sunlight")
plt.title("rainfall vs ph")
plt.show()
```

```python
df.head(1000).plot(kind='scatter',x='temperature',y='humidity',title="Temperature vs Humidity",color="green")
```

```python
df.plot(kind='scatter',x='rainfall',y='crop',title="Rainfall vs Crop_predicted")
```

```python
df['temperature'].plot(kind='hist', bins=20,
    color='green',
    edgecolor='red',
    figsize=(6,4),
      xlabel="Temperature (°C)",
    title="Temperature Distribution")
```

```python
df[['temperature','humidity','rainfall']].plot(kind='box',figsize=(8,5),color=dict(boxes='blue',whiskers='green',medians='red',caps='black'),title='')
```

```python
df.plot(kind='scatter',
        x='ph',
        y='crop',
       alpha=0.1,
        color='maroon',
        s=50,
        figsize=(6,4),
        title="Soil_pH vs Crop_Predicted")
```

```python
df[['nitrogen','phosphorus','potassium']].mean().plot(kind='pie',autopct='1.1f%%',title='NPK Average')
plt.ylabel('')
plt.show()
```

```python
crop_counts=df['crop'].value_counts()
crop_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    figsize=(6,6),
    title='Crop Distribution')
plt.ylabel('')
plt.show()
```

```python
df.mean().plot(
    kind='bar',
    figsize=(10,5),
    color='cyan',
    edgecolor='black',
    title='Analysis of Environmental and Soil Factors Influencing Crop Production'
)
```

```python
df[['temperature','rainfall']].head(100).plot(kind='area',figsize=(7,4),title='Temperature vs Rainfall',alpha=0.6)
plt.show()
```

```python
df['crop'].value_counts().plot(
    kind='barh',
    color='teal',
    figsize=(7,5),
    title='Horizontal Bar chart:Crop production'
)
plt.show()
```

```python
df[['temperature','humidity','rainfall']].plot(kind='kde')
plt.show()
```

```python
df[['temperature','humidity','rainfall']].head(20).plot(
    kind='bar',
    stacked=True,
    figsize=(10,5),
    title='Stacked Weather Variables'
)
```

```python
import seaborn as sns
corr=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Between Crop Prediction Factors')
plt.show()
```
